```python
import genetic
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

desired_answer = 0.003224
accuracy = 9999
max_generations = 1000

gen_alg = genetic.Genetic_Algorithm(desired_answer, accuracy, max_generations)
start_population = gen_alg.get_random_generation(gen_size=1000, random_boundary=1000)
best_score, best_ancestors, best_results = gen_alg.genetic_algorithm(start_population)
best_ancestor = gen_alg.get_best_ancestor(best_score[0], best_ancestors)
received_answer = gen_alg.test_fitness(best_ancestor[0],best_ancestor[1],best_ancestor[2])

print(f'Desired value = {desired_answer}\tGA received value = {received_answer}')
print(f'Best generation: {int(best_score[0])}/{max_generations}')
print("Highest score: ", best_score[1])
print("Most appropriete function values: ", best_ancestor)
print('Fitness value of the best solution =', abs(desired_answer - received_answer))
print('Accuracy:', gen_alg.get_accuracy(best_ancestor))
```
